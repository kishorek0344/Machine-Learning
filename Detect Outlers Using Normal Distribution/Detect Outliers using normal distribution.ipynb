```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("heights_with_outliers.csv")
```

```python
df.head(3)
```

```python
df.shape
```

```python
mean = df.height.mean()
std=df.height.std()
mean,std
```

```python
sns.histplot(df.height ,kde=True)
plt.axvline(mean, color="black", linestyle="dashed", linewidth='1')

#one std
plt.axvline(mean+std, color="r", linestyle="dashed", linewidth='1')
plt.axvline(mean-std, color="r", linestyle="dashed", linewidth='1')

#two std
plt.axvline(mean+2*std, color="g", linestyle="dashed", linewidth='1')
plt.axvline(mean-2*std, color="g", linestyle="dashed", linewidth='1')

#three std
plt.axvline(mean+3*std, color="b", linestyle="dashed", linewidth='1')
plt.axvline(mean-3*std, color="b", linestyle="dashed", linewidth='1')

```

```python
left3=mean-3*std
right3=mean+3*std
left3,right3
```

```python
df[(df.height < left3) | (df.height > right3)]
```

```python
no_outliers=df[(df.height > left3) & (df.height < right3)]
no_outliers.shape
```

```python
sns.histplot(no_outliers.height, kde=True)
```
