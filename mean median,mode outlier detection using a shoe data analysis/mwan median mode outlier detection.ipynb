## ADDIDAS Analysis

```python
df_addidas = df[df.brand=='Adidas']
df_addidas.head()
```

```python
def plot_qty():
    plt.figure(figsize=(12,5))
    dates = df_nike.date
    plt.plot(dates, df_nike.sold_qty,marker='o', label='Nike', color='blue')
    plt.plot(dates, df_addidas.sold_qty, marker='+', label='Addidas', color='red')
    plt.xlabel("Dates")
    plt.ylabel("shoes sold quantity")
    plt.title("SHOE ANALYSIS")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
```

```python
plot_qty()
```

```python
df_addidas.replace(689, vall, inplace=True)
```
